# Data Preparation

This notebook finds files with encoding errors in 1_raw_txt and sends to 1_raw_txt_error.

Files with encoding errors are not used in the study.

```python
import os

from google.colab import drive
drive.mount('/content/drive')
```

```python
# file encoding error, identified by 'ÿ'
project_path = '/content/drive/MyDrive/Monetary Policy Coordination'

file_path = f'{project_path}/Data/1_raw_txt'
error_path = f'{project_path}/Data/1_raw_txt_error'
```

```python
error_count = 0

for cb in os.listdir(file_path):
  for file_name in os.listdir(os.path.join(file_path, cb)):
    with open(os.path.join(file_path, cb, file_name),
              'r',
              errors='ignore') as file:
      if 'ÿ' in file.read():
        error_count += 1
        if not os.path.exists(os.path.join(error_path, cb)):
          os.makedirs(os.path.join(error_path, cb))
        os.rename(os.path.join(file_path, cb, file_name),
                  os.path.join(error_path, cb, file_name))
error_count
```
